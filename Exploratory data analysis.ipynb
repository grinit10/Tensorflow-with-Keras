```python
df_without_top_one_percent = df.sort_values('price', ascending=False).iloc[216:]
df_without_top_one_percent.head()
```

```python
df_without_top_one_percent.drop('id', axis=1, inplace=True)
```

```python
df_without_top_one_percent['date'] = pd.to_datetime(df_without_top_one_percent['date'])
```

```python
df_without_top_one_percent['year'] = df_without_top_one_percent['date'].apply(lambda date: date.year)
df_without_top_one_percent['month'] = df_without_top_one_percent['date'].apply(lambda date: date.month)
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='month', y='price', data=df_without_top_one_percent);
```

```python
df_without_top_one_percent.groupby('month').mean()['price'].plot();
```

```python
df_without_top_one_percent.groupby('year').mean()['price'].plot();
```

```python
df_without_top_one_percent.drop('date', inplace=True, axis=1)
```

```python
df_without_top_one_percent.drop('zipcode', inplace=True, axis=1)
```

```python
df_without_top_one_percent[df_without_top_one_percent['yr_renovated'] > 0].groupby('yr_renovated').mean()['price'].plot();
```

```python
X = df_without_top_one_percent.drop('price', axis=1).values
y = df_without_top_one_percent['price'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit_transform(X_train)
scalar.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss='mse')
```

```python
model.fit(x = X_train, y = y_train, batch_size = 128, validation_data=(X_test, y_test), epochs=400)
```

```python
pd.DataFrame(data = model.history.history).plot();
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
y_pred = model.predict(X_test)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
regression_df = pd.DataFrame(data=y_pred, columns=['Predicted Y'])
regression_df['Actual Y'] = pd.DataFrame(data=y_test)
regression_df
```
