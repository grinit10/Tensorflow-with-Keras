```python
import seaborn as sns
```

```python
heart_csv = pd.read_csv('heart.csv')
heart_csv.head()
```

```python
sns.distplot(heart_csv['age']);
```

```python
plt.figure(figsize=(12, 12))
sns.distplot(heart_csv['age'], kde=False);
```

```python
plt.figure(figsize=(12, 12))
sns.distplot(heart_csv['age'], kde=False, bins=40,color='red');
```

```python
sns.countplot(x='sex',data=heart_csv);
```

```python
sns.countplot(x='target',data=heart_csv);
```

```python
sns.countplot(x='cp',data=heart_csv);
```

```python
sns.countplot(x='cp',data=heart_csv,hue='sex');
```

```python
sns.countplot(x='cp',data=heart_csv,palette='terrain');
```
